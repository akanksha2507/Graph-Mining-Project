```python
import os, operator, nltk
from collections import defaultdict
from nltk.corpus import stopwords
from nltk.tag import pos_tag
import networkx as nx
from nltk.stem.snowball import SnowballStemmer
import string
```

```python
stemmer = SnowballStemmer("english")
```

```python
def computeMetric(origKeys, predKeys):
    n_hits = 0
    n_misses = 0
    n_false_alarms = 0
    for k in predKeys:
        if k in origKeys:
            n_hits += 1
        else:
            n_false_alarms += 1
    for k in origKeys:
        if k not in predKeys:
            n_misses += 1
    if(n_hits==0):
        precision=0
        recall=0
        F1_score=0
    else:
        precision = n_hits*1.0/(n_hits+n_false_alarms)
        recall = n_hits*1.0/(n_hits+n_misses)
        F1_score = 2.0*(precision*recall)/(precision+recall)
    return precision,recall,F1_score
```

```python
nodes={}
dataset_dir = '/Users/akankshagrover/Desktop/celf++_code_release/data/Hulth_prof_7/map'
for subdir, dirs, files in os.walk(dataset_dir):
    for file in files:
        f = open(subdir+'/'+file, 'r')
        lines = f.readlines()
        fileName=file[4:].split(".")
        node_dict={}
        for line in lines:
            line=line.split()
            node_dict[line[0]]=line[1].strip('"') 
        nodes[fileName[0]]=node_dict
```

```python
dataset_dir = '/Users/akankshagrover/Desktop/celf++_code_release/new_outputs/Hulth_7_test'
file_dict={}  
for subdir, dirs, files in os.walk(dataset_dir):
    for file in files:
        if not file.endswith('.abstr'):
            continue
        f = open(subdir+'/'+file, 'r')
        score_list=[]
        for line in f:
            line=line.split()
            score_list.append((line[0],float(line[2])))
        score_list.sort(key=lambda tup: tup[1],reverse=True)
        keywords=[]
        file=file[7:]
        file=file.split('.')
        node_key=nodes[file[0]]
        for k,v in score_list:
            keywords.append(node_key[k])
        #keywords_new=" ".join(keywords)
        exclude = set(string.punctuation)
        s = ' '.join(ch for ch in keywords if ch not in exclude)
        s=s.split(' ')
        file_dict[file[0]]= s
```

```python
org_dict={}
dataset_dir = '/Users/akankshagrover/Desktop/celf++_code_release/data/Hulth_prof_7/keywords'
for subdir, dirs, files in os.walk(dataset_dir):
    for file in files:
        f = open(subdir+'/'+file, 'r')
        lines = f.readlines()
        origKeys=[]
        for line in lines[1:]:
            line=line.split()
            origKeys.append(line[1].strip('"'))
        origKeys=[stemmer.stem(t) for t in origKeys]
        fileName=file[9:].split(".")
        org_dict[fileName[0]]=origKeys
```

```python
precision =[]
recall=[]
f1=[]
for k in file_dict.keys():
    pred_keys=file_dict[k]
    org_keys=org_dict[k]
    print "predicted", len(pred_keys)
    #print pred_keys
    print"original",len(org_keys)
    #print org_keys
    p,r,f=computeMetric(org_keys,pred_keys)
    precision.append(p)
    recall.append(r)
    f1.append(f)
```
